```python
import numpy as np
```

```python
np.random.seed(123)  #ensures that the same random numbers are generated every time.
random_walk=[0]      #we are starting at step 0
```

```python
for i in range(100): #simulating 100 steps
    step=random_walk[-1] #the step you are at is the very last in the random walk list (-1)

    #now simulate the dice roll
    dice=np.random.randint(1,7) #you want the dice to generate random integers on a dice
    if dice<=2:
        step=step-1 #1 or 2, back one step
        #to avoid going below 0, step=max(0,step-1)
    elif dice<=5:
        step=step+1  #3,4 or 5 forward one step
    else:
        step=step+np.random.randint(1,7)

    random_walk.append(step) #add the new position to the random walk list

print(random_walk)
```

```python
import matplotlib.pyplot as plt

plt.plot(random_walk)
plt.title("Random Walk ")
plt.xlabel("Step Number")
plt.ylabel("Position")
plt.grid(True)
plt.show()
```
