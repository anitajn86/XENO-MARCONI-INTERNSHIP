Using word2Vec

```python
#Train Word2Vec on Tokenized Questions
from gensim.models import Word2Vec
```

```python
w2v_model = Word2Vec(sentences=xeno['question_tokens'], vector_size=100, window=5, min_count=1,workers=4,sg=1)
```

```python
#Get sentence/question embeddings
import numpy as np
```

```python
def sentence_vector(tokens,model):
    vectors=[model.wv[word] for word in tokens if word in model.wv]
    if not vectors:
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis=0)
```

```python
xeno['question_vector'] = xeno['question_tokens'].apply(lambda tokens: sentence_vector(tokens, w2v_model))
```

```python
#Convert the vectors to a matrix
question_matrix = np.vstack(xeno['question_vector'].values)
```

```python
#Embed a new query and compute its similarity
from sklearn.metrics.pairwise import cosine_similarity
from nltk.tokenize import regexp_tokenize
```

```python
# Tokenize function
def tokenize(text):
    pattern = r"\b[a-zA-Z]+\b"  # match only alphabetic words
    return regexp_tokenize(text.lower(), pattern)
```

```python
# Semantic search function
def semantic_search(query, model, data, top_n=5):
    # Tokenize and vectorize query
    query_tokens = tokenize(query)
    query_vector = sentence_vector(query_tokens, model).reshape(1, -1)

    # Compute cosine similarities
    similarities = cosine_similarity(query_vector, question_matrix)[0]

    # Get top N results
    top_indices = similarities.argsort()[-top_n:][::-1]

    results = data.iloc[top_indices][['question', 'content']]
    results['similarity'] = similarities[top_indices]
    return results.reset_index(drop=True)
```

```python
def semantic_search(query, model, data, top_n=5, threshold=0.1):
    query_tokens = tokenize(query)
    query_vector = sentence_vector(query_tokens, model).reshape(1, -1)

    similarities = cosine_similarity(query_vector, question_matrix)[0]
    top_indices = similarities.argsort()[::-1]

    results = []
    for idx in top_indices:
        if similarities[idx] < threshold:
            break
        results.append({
            'question': data.iloc[idx]['question'],
            'content': data.iloc[idx]['content'],
            'similarity': similarities[idx]
        })
        if len(results) == top_n:
            break
    return pd.DataFrame(results)
```

```python
results = semantic_search("XENO and autosave accounts", w2v_model, xeno)
print(results)
```

Using sentence Transformer

```python
from sentence_transformers import SentenceTransformer, util
import torch
```

```python
model = SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')  #
#'all-MiniLM-L6-v2  Lightweight, fast, good performance
```

```python
# Your list of questions
questions = xeno['question'].tolist()

# Encode all at once into a tensor matrix
question_embeddings = model.encode(questions, convert_to_tensor=True)
```

```python
query = "XENO and autosave accounts"
query_embedding = model.encode(query, convert_to_tensor=True)

# Compute cosine similarities
cosine_scores = util.cos_sim(query_embedding, question_embeddings)[0]  # shape: [num_questions]
```

```python
# Convert to numpy
scores = cosine_scores.cpu().numpy()

# Get top N results
top_n = 5
top_indices = scores.argsort()[-top_n:][::-1]

# Prepare result DataFrame
results = xeno.iloc[top_indices][['question', 'content']].copy()
results['similarity'] = scores[top_indices]

print(results)
```
