```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("XENO'S DATA.xlsx")
df
```

```python
df.head()
```

```python
#Number of questions per category
qns_category=df['CATEGORY'].value_counts()
print(qns_category)
```

```python
df.isna().any()
```

```python
#Question length in characters
df['character_length']=df['QUESTION'].str.len()
```

```python
#Question length in words
df['no_of_words']=df['QUESTION'].str.split().str.len()
```

```python
df.head()
```

```python
#Average  question length in character by category
character_ave=df.groupby('CATEGORY')['character_length'].mean()
character_ave
```

```python
#Average  question length in word by category
word_ave=df.groupby('CATEGORY')['no_of_words'].mean()
word_ave
```
